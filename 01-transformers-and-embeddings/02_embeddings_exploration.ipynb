```python
from transformers import AutoTokenizer, AutoModel
import torch

print("Import successful!")
```

```python
# Model load karte hain
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
model = AutoModel.from_pretrained("distilbert-base-uncased")

print("Model ready!")
```

```python
# Do words ko embed karte hain
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    # CLS token ki embedding = poore sentence ka representation
    embedding = outputs.last_hidden_state[:, 0, :]
    return embedding

# Teen words try karte hain
king = get_embedding("king")
queen = get_embedding("queen")
apple = get_embedding("apple")

print("King embedding shape:", king.shape)
```

```python
from torch.nn.functional import cosine_similarity

# Similarity nikaalte hain
king_queen = cosine_similarity(king, queen)
king_apple = cosine_similarity(king, apple)
queen_apple = cosine_similarity(queen, apple)

print(f"King — Queen similarity: {king_queen.item():.4f}")
print(f"King — Apple similarity: {king_apple.item():.4f}")
print(f"Queen — Apple similarity: {queen_apple.item():.4f}")
```

```python
sentence1 = get_embedding("the king ruled the kingdom")
sentence2 = get_embedding("the queen ruled the kingdom")
sentence3 = get_embedding("I ate a fresh apple today")

sim_1_2 = cosine_similarity(sentence1, sentence2)
sim_1_3 = cosine_similarity(sentence1, sentence3)

print(f"King sentence — Queen sentence: {sim_1_2.item():.4f}")
print(f"King sentence — Apple sentence: {sim_1_3.item():.4f}")
```

```python
s1 = get_embedding("This vaccine causes serious side effects")
s2 = get_embedding("This vaccine leads to harmful reactions")
s3 = get_embedding("The stock market crashed today")

sim_12 = cosine_similarity(s1, s2)
sim_13 = cosine_similarity(s1, s3)

print(f"Similar claim similarity: {sim_12.item():.4f}")
print(f"Different topic similarity: {sim_13.item():.4f}")
```
